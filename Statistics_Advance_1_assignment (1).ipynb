# **Theory**


 1.What is a random variable in probability theory?
ANS:A random variable is a variable that takes on different numerical values based on the outcome of a random experiment.

 2.What are the types of random variables?
ANS:Discrete random variable: Takes countable values.
Continuous random variable: Takes any value in a range.

 3.What is the difference between discrete and continuous distributions?
ANS:Discrete: Finite/countable outcomes (e.g., coin toss).
Continuous: Infinite outcomes in a range (e.g., height).

 4.What are probability distribution functions (PDF)?
ANS:PDF describes the likelihood of a random variable taking a specific value

 5.How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
ANS:CDF gives the probability that a variable is less than or equal to a value, while PDF gives the likelihood at an exact value

 6.What is a discrete uniform distribution?
ANS:It’s a distribution where all outcomes are equally likely

 7.What are the key properties of a Bernoulli distribution?
ANS:Two outcomes: Success (1) or Failure (0)
One trial
Probability of success = p

 8.What is the binomial distribution, and how is it used in probability?
ANS:It models the number of successes in n independent Bernoulli trials (e.g., 5 coin tosses).

 9.What is the Poisson distribution and where is it applied?
ANS:It models the number of events occurring in a fixed interval (e.g., number of emails per hour), especially when events are rare.

 10.What is a continuous uniform distribution?
ANS:All values in a range have equal probability (e.g., selecting a random number between 0 and 1).

 11.What are the characteristics of a normal distribution?
ANS:Bell-shaped curve
Symmetric around the mean
Mean = Median = Mode

 12.What is the standard normal distribution, and why is it important?
 ANS:A normal distribution with mean 0 and standard deviation 1; it’s used to calculate Z-scores and compare different normal distributions.

 13.What is the Central Limit Theorem (CLT), and why is it critical in statistics?
ANS:CLT states that the sampling distribution of the sample mean becomes approximately normal as sample size increases, regardless of the population distribution.

 14.How does the Central Limit Theorem relate to the normal distribution?
ANS:It explains why the normal distribution is common in statistics—even when the original data isn’t normal.

 15.What is the application of Z statistics in hypothesis testing?
ANS:Z-statistics are used to test hypotheses when the population standard deviation is known, especially in large samples.

 16.How do you calculate a Z-score, and what does it represent?
ANS:Z = (X - μ) / σ
It measures how many standard deviations a value (X) is from the mean (μ).

 17.What are point estimates and interval estimates in statistics?
ANS:Point estimate: A single value (e.g., sample mean)
Interval estimate: A range (e.g., confidence interval) to estimate the parameter

 18.What is the significance of confidence intervals in statistical analysis?
ANS:They show the range in which the true population parameter likely falls, with a certain confidence level (e.g., 95%).

 19.What is the relationship between a Z-score and a confidence interval?
ANS:Z-scores define the boundaries of confidence intervals (e.g., Z = ±1.96 for 95% CI).

 20.How are Z-scores used to compare different distributions?
ANS:They standardize values, allowing comparison across different distributions or units.

 21.What are the assumptions for applying the Central Limit Theorem?
ANS:Samples are independent
Sample size is large (n ≥ 30)
Population variance is finite

 22.What is the concept of expected value in a probability distribution?
ANS:It is the average or mean value expected if the experiment is repeated many times.

 23.How does a probability distribution relate to the expected outcome of a random variable?
ANS:It helps compute the expected value by weighting each possible outcome with its probability.

```python
 #Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution
data_point = 75
mean, std = 70, 5
z = (data_point - mean) / std
print("Z-score:", z)
print("Probability from Z:", norm.cdf(z))
```

```python
 #Implement hypothesis testing using Z-statistics for a sample dataset
sample = np.random.normal(50, 10, 30)
sample_mean = np.mean(sample)
pop_mean = 52
z_stat = (sample_mean - pop_mean) / (np.std(sample, ddof=1) / np.sqrt(len(sample)))
p_value = 2 * (1 - norm.cdf(abs(z_stat)))
print("Z-stat:", z_stat, "P-value:", p_value)
```

```python
 #Create a confidence interval for a dataset using Python and interpret the result
conf_int = stats.norm.interval(0.95, loc=sample_mean, scale=stats.sem(sample))
print("95% Confidence Interval:", conf_int)
```

```python
 #Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean
normal_data = np.random.normal(60, 8, 100)
mean = np.mean(normal_data)
ci = stats.norm.interval(0.95, loc=mean, scale=stats.sem(normal_data))
print("CI for mean of normal data:", ci)
```

```python
# Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution
data = np.random.normal(100, 15, 100)
data_range = np.ptp(data)
iqr = stats.iqr(data)
print("Range:", data_range, "IQR:", iqr)
```

```python
# Implement Z-score normalization on a dataset and visualize its transformation
data = np.random.normal(60, 12, 100)
z_normalized = zscore(data)
sns.histplot(z_normalized, kde=True)
plt.title("Z-score Normalization")
plt.show()
```
